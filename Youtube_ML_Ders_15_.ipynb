```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
#birden fazla treee kuruyor ve bu ağaçlar arasında karar veriyr.
#aşırı öğrenmede direnç gösterir
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("insurance.csv")
df.head(3)
```

```python
df = pd.get_dummies(df,columns=["sex","smoker","region"],drop_first=True, dtype=int)
```

```python
df.head(3)
```

```python
y=df["charges"]
x=df.drop("charges",axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=22,train_size=0.7)
```

```python
lr=LinearRegression()
model = lr.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
rf=RandomForestRegressor(n_estimators=200)#n_estimators ağaç sayısı hiç bir şey girilmezse 100 olur
#max_depth ağacın ineceği derinlik
model = rf.fit(x_train,y_train)
model.score(x_test,y_test)
```
