```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error,mean_absolute_percentage_error
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head(3)
```

```python
df = pd.get_dummies(df, columns=['sex', 'smoker', 'region'], drop_first=True, dtype=int)
```

```python
df.head(3)
```

```python
y = df['charges'] # y tahmin etmek istediğimiz target
x = df.drop('charges', axis=1) # x feature özellik
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[18,26,1,0,1,1,0,1]])
```

```python
df_hata = pd.DataFrame()
```

```python
df_hata['y']=y
```

```python
df_hata
```

```python
model.predict(x) #elimizdeki x leri veriyoruz y yi tahmin etmesini istiyoruz
```

```python
y_tahmin = model.predict(x)
```

```python
df_hata['y_tahmin'] = y_tahmin
```

```python
df_hata #gerçek y , tahmin y
```

```python
df_hata['error'] = y_tahmin
```

```python
df_hata.head(10)
```

```python
#MSE   #hata karelerin ortalaması
```

```python
df_hata['squared_error'] = df_hata['error']**2
```

```python
df_hata
```

```python
#MAE (Mutlak değer -Absolute value)
```

```python
df_hata['abs_error'] = np.abs(df_hata['error'])
```

```python
df_hata
```

```python
#Mape
```

```python
df_hata['percent_error'] = np.abs(y-y_tahmin)/y
```

```python
df_hata
```

```python
df_hata.mean()
```
