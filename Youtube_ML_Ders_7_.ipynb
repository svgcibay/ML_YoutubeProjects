```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset("diamonds")
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['cut', 'color', 'clarity'], drop_first=True, dtype=int)
```

```python
df.head()
```

```python
y = df["price"]
x = df.drop("price", axis=1)
```

```python
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size=0.76,random_state=19)
```

```python
lm=LinearRegression()
model=lm.fit(x_train,y_train)
```

```python
# fark ne kadar azsa o kadar uyumlu öğrenme olmuştur
# fark ne kadar fazla ise aşırı öğrenme olmuştur -OVERFITTING
# The smaller the difference, the more adaptive learning has been
# The greater the difference, the more overlearning has occurred -OVERFITTING
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
