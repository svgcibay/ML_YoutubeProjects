```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('audi.csv')
```

```python
df.head()
```

```python
df=df[["Year","Type","Mileage(miles)","Engine","PS","Transmission","Fuel","Number_of_Owners","Price(£)"]]
```

```python
df.head(3)
```

```python
# y= ax+b
```

```python
df.columns=["yil","kasa","mil","motor","ps","vites","yakit","sahip","fiyat"]
```

```python
df["motor"]=df["motor"].str.replace("L","")
df["motor"]=pd.to_numeric(df["motor"])
```

```python
df=pd.get_dummies(df,columns=["kasa","vites","yakit"],drop_first=True)
```

```python
df.head(3)
```

```python
y=df[["fiyat"]]     #fiyat sütunun y ye ata y bağımlı değişkendir !
x=df.drop("fiyat",axis=1)   #fiyat sütununu cıkar diğer herşeyi x değişkenine ata
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=22)
```

```python
lm=LinearRegression()
model=lm.fit(x_train,y_train)
model.score(x_test,y_test)
```
